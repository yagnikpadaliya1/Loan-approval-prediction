```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
data = pd.read_csv("loan_approval_data.csv")
data.sample(5)
```

```python
data.info()
```

Handling missing values

```python
# Handling missing data

cat_col = data.select_dtypes(include=["object"]).columns # this is hepls to combine the all categorycal data
num_col = data.select_dtypes(include=["number"]).columns # this is hepls to combine the all numerical data

#num_col = data.select_dtypes(include=["float64" , "int64"]).columns 
# this is also one type to declare the dtypes
```

```python
from sklearn.impute import SimpleImputer

num_impl = SimpleImputer(strategy="mean")
data[num_col] = num_impl.fit_transform(data[num_col])

cat_impl = SimpleImputer(strategy="most_frequent")
data[cat_col] = cat_impl.fit_transform(data[cat_col])
```

```python
data.head()
```

```python
x = data.drop(columns = ["Applicant_ID","Loan_Approved"])
y = data["Loan_Approved"]
```

EDA


```python
# Transform our data

x = pd.get_dummies(x,columns = ["Employment_Status","Loan_Purpose","Property_Area","Employer_Category"] , drop_first = True , dtype = int)

x["Gender"] = x["Gender"].map({"Female" : 0 ,"Male" : 1})
x["Marital_Status"] = x["Marital_Status"].map({"Single" : 0 ,"Married" : 1})
x["Education_Level"] = x["Education_Level"].map({"Not Graduate" : 0 ,"Graduate" : 1})

y["Loan_Approved"] = y["Loan_Approved"].map({"No" : 0 ,"Yes" : 1})

# x["cp_hartrate"] = x["cp"] * x["thalach"]
```

```python
x.sample(5)
```

```python
x.info()
```

```python
# Train_test_split

x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
# Logistic Regration

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 500)
model.fit(X_train_scaled,y_train)

y_pred = model.predict(X_test_scaled)
```

```python
print("Logistic Regression Model")
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1 score: ", f1_score(y_test, y_pred))
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("CM: ", confusion_matrix(y_test, y_pred))
```
